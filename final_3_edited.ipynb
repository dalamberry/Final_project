
Идеи для метрик, по которым можно оценивать события в игре можно разделить на две основные группы: рост(то, что поможет расшириться и заработать) и продукт(то, что поможет улучшить сам продукт)

1. **РОСТ**. Количество игроков, принявших участие в событии
2. **РОСТ**. Процент игроков от активных, принявших участие в событии
3. **РОСТ**. Среднее время нахождения в событии (в день и распределение по дням)
4. **ПРОДУКТ**. Сколько и каких призов было получено в общем и по пользователям
5. **ПРОДУКТ**. Среднее время, требующееся для прохождения уровня
6. **ПРОДУКТ**. Среднее количество попыток  пройти уровень
7. **ПРОДУКТ**. Процент пользователей, которые смогли пройти уровень до конца

Из существующих метрик - ER(коэффициент вовлеченности), Sessions per user(частота использования),
Key user actions per session(частота ключевых действий за сессию), Retention rate(возвращение игроков), CR(конверсия), 
DAU, WAU, MAU(количество уникальных пользователей за день, неделю, месяц).


В случае усложнения механики игры можно предположить, что набор метрик станет
более склонным к корреляциям и взаимозависимостям. Кроме идей для метрик, предложенных выше,
нужно искать связи, как то:

1. **ПРОДУКТ**. Ключевая метрика:  какой процент пользователей доходит и до каких уровней 
2. **РОСТ**. Отследить, есть ли изменения в среднем времени провождения в игре
3. **ПРОДУКТ**. Отследить, меняется ли количество попыток пройти уровень
4. **ПРОДУКТ**. Поиск корреляции между неудачными попытками и другими метриками

И также более внимательно отслеживать вовлеченность пользователей.
